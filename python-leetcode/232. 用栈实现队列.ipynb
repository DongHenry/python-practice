#### 题目描述
使用栈实现队列的下列操作：

push(x) -- 将一个元素放入队列的尾部。

pop() -- 从队列首部移除元素。

peek() -- 返回队列首部的元素。

empty() -- 返回队列是否为空。

---
说明:

你只能使用标准的栈操作 -- 也就是只有 push to top, peek/pop from top, size, 和 is empty 操作是合法的。
你所使用的语言也许不支持栈。你可以使用 list 或者 deque（双端队列）来模拟一个栈，只要是标准的栈操作即可。
假设所有操作都是有效的 （例如，一个空的队列不会调用 pop 或者 peek 操作）。

#### Solution1
使用Python队列实现

- 个人认为使用python队列实现这个方法有误。 原题要求用栈来实现队列，而下面的实现中直接用到了queue.insert(0, x) 和queue.pop()等方法，这些方法都是队列的方法。其根本是用队列实现队列，而不是用栈实现队列。

##### 执行结果
- 执行用时 :20 ms, 在所有 Python 提交中击败了90.56%的用户
- 内存消耗 :11.9 MB, 在所有 Python 提交中击败了8.56%的用户

```python
class MyQueue(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        # python中队列用列表表示
        self.queue = []
        

    def push(self, x):
        """
        Push element x to the back of queue.
        :type x: int
        :rtype: None
        """
        # 向队尾插入元素，入队
        self.queue.insert(0, x)

    def pop(self):
        """
        Removes the element from in front of queue and returns that element.
        :rtype: int
        """
        # 队头元素出队
        return self.queue.pop()
        

    def peek(self):
        """
        Get the front element.
        :rtype: int
        """
        # 返回队头元素
        return self.queue[-1]
        

    def empty(self):
        """
        Returns whether the queue is empty.
        :rtype: bool
        """
        return False if self.queue else True
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

#### Solution 2 正解
用双栈来实现队列，一个栈为输入栈，一个栈为输出栈
- 输入栈用于实现数据元素入队的操作push(x)
- 输出栈用于实现数据元素出队pop()和查看队首元素peek()的操作
- push(x) 将x添加到输入栈，此操作相当于入队
- pop() 从输出栈栈顶取走元素，相当于出队,若输出栈没有元素，将输入栈中的元素一次出栈压入输出栈，再从输出栈取走栈顶元素
- peek() 与pop原理相同
- 输入栈和输出栈全为空，队列为空

---- 
##### 执行结果
- 执行用时 :52 ms, 在所有 Python3 提交中击败了61.47%的用户
- 内存消耗 :12.9 MB, 在所有 Python3 提交中击败了98.48%的用户

```python
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.input = []
        self.output = []
        

    def push(self, x: int) -> None:
        """
        Push element x to the back of queue.
        """
        self.input.append(x)
        

    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        if not self.output:
            while self.input:
                a = self.input.pop()
                self.output.append(a)
        return self.output.pop()
        

    def peek(self) -> int:
        """
        Get the front element.
        """
        if not self.output:
            while self.input:
                a = self.input.pop()
                self.output.append(a)
        return self.output[-1]
        

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        if not self.input and not self.output:
            return True
        else:
            return False
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
