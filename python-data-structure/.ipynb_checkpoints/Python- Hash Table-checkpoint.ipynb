Hash tables are a type of data structure in which the address or the index value of the data element is generated from a hash function . In others words hash table stores key-values pairs but the key is generated through a hashing function

In Python, the Dictionary data types represent the implementation of hash tables. The Keys in the dictionary satisfy the following requirements.

- The keys of the dictionary are hashable i.e. the are generated by hashing function which generates unique result for each unique value supplied to the hash function.
- The order of data elements in a dictionary is not fixed.

#### Accessing values in dictionary

```python
# declare a dictionary
info = {'name': 'dong', 'age': 21, 'country': 'china'}

# accessing the dictionary with the key
print(f"His name is : {info['name']}")
```

#### Updating dictionary

```python
# update existing entry
info['age'] = 18

# add new entry
info['school'] = 'CIDP'
info
```

#### Deleting dictionary elements

```python
# remove entry with key 'name'
del info['name']
print(info)

# remove all entries in dict
# info.clear()
```
